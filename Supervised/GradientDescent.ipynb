```python
import numpy as np
```

```python
def gradient(x,y):
    m_curr = b_curr = 0
    iterations = 1000
    if len(x) != len(y):
        print ('Error!! length of the two arrays should be equal.')
    n = len(x)
    eta = 0.001
    
    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        md = -(2/n)*sum(x*(y-y_predicted))
        bd = -(2/n)*sum(y-y_predicted)
        m_curr = m_curr - eta*md
        b_curr = b_curr - eta*bd
        print (f'm {m_curr}, b {b_curr}, cost {cost}, step {i}')
```

```python
x=np.array([10,20,45,19])
y=np.array([1,0,1,0])
```

```python
gradient(x,y)
```
